```python
gdpdata=gdpdata.fillna(0)
```

```python
def country():
  n=int(input("Enter the number of years for analysis:"))
  names = gdpdata['Country Name'].unique()
  print(names)
  x = str(input("Enter Country Name for analysis: "))
  data1 = gdpdata[gdpdata['Country Name']==x]
  display(data1)
  data2 = data1.iloc[:,-n:]
  display(data2)
  data2.plot.bar(xlabel=[x],ylabel=['GDP of the country'])
  plt.show()
```

```python
country()
```

```python
def top_10_countries():
  choice=int(input("Enter '1' to get details about top 10 countries or enter '2' to get details about bottom 10 countries: "))
  year=int(input("Enter the year of comparision between the countries: "))
  if(choice==1):
    a=int(input("Enter the number of countries that you want to print"))
    data1 = pd.DataFrame()
    data1 = gdpdata[['Country Name',str(year)]]
    data1 = data1.sort_values(by=str(year),ascending=False)
    data2=data1.head(a+1)
    display(data2)
    data2 = data2.groupby(['Country Name'],as_index=False).sum()
    data2.plot.bar(x='Country Name',y=str(year),xlabel='Country Name')
    plt.show()
    plt.pie(data2[str(year)],labels=data2['Country Name'])
  elif(choice==2):
    b=int(input("Enter the number of countries that you want to print: "))
    data1 = pd.DataFrame()
    data1 = gdpdata[['Country Name',str(year)]]
    data1 = data1.sort_values(by=str(year),ascending=True)
    data2=data1.head(b)
    display(data2)
    data2 = data2.groupby(['Country Name'],as_index=False).sum()
    data2.plot.bar(x='Country Name',y=str(year),xlabel='Country Name')
    plt.show()
    plt.pie(data2[str(year)],labels=data2['Country Name'])
  else:
    print("Kindliy enter a valid choice")
```

```python
top_10_countries()
```

```python
def percentage():
  num=int(input("Enter the number of countries with which you want to compare with rest of the world "))
  year = int(input("Enter year of comparison: "))
  country=str(input("Enter the name of coutries with which you want to compare with rest of the world "))
  data1 = pd.DataFrame()
  data1 = gdpdata[['Country Name',str(year)]]
  data1 = data1.sort_values(by=str(year),ascending=False)
  data1 = data1.head(num+1)
  display(data1)
  a = data1[[str(year)]]
  a = pd.DataFrame(a.sum())
  ans = a._get_value(str(year),0)
  b=gdpdata[gdpdata['Country Name']=='World']
  b = b[[str(year)]]
  d = b._get_value(189,str(year))
  print(d)
  ans = ans - d
  d=d-ans
  list1 = [ans, d]
  plt.pie(list1,labels = ['Top Countries','Rest of the world'],explode=(.1,0),colors = ("Yellow","Blue"))
  plt.show()
  plt.bar(x=['Top Countries','Rest of the world'],height=[ans, d],width=0.2,color=("Pink","Blue"))
  plt.xlabel('Group Name')
  plt.ylabel('Size')
  plt.show()
  data3=gdpdata[gdpdata['Country Name']==country]
  data3=data3[['Country Name',str(year)]]
  display(data3)
  data4=gdpdata[gdpdata['Country Name']=='World']
  data4=data4[['Country Name',str(year)]]
  display(data4)
  data5=[data3[str(year)],data4[str(year)]]
  print(data5)
  plt.pie(data5,labels=[str(country),'World'],colors=("Green","Red"))
```

```python
percentage()
```

```python
def gdp_growth():
  print(gdpdata['Country Name'].unique())
  c=str(input("Enter country name: "))
  y=int(input("Enter year: "))
  data1 = gdpdata[gdpdata['Country Name']==c]
  data1 = data1[['Country Name',str(y)]]
  display(data1)
  data2 = gdpdata[gdpdata['Country Name']==c]
  data2=data2[['Country Name',str(y-1)]]
  display(data2)
  new = pd.DataFrame()
  new['Proportion'] =((data1[str(y)]-data2[str(y-1)])/data1[str(y)])*100
  display(new)
```

```python
gdp_growth()
```

```python
def country_comparision():
  print(gdpdata['Country Name'].unique())
  country1=str(input("Enter the name of country 1:"))
  country2=str(input("Enter the name of country 2:"))
  year=int(input("Enter the year of comaprision:"))
  value1=gdpdata[gdpdata['Country Name']==country1]
  value1=value1[[str(year)]]
  value1['temp'] = year
  display(value1)
  a=value1[str(year)]
  value2=gdpdata[gdpdata['Country Name']==country2]
  value2=value2[[str(year)]]
  value2['temp'] = year
  display(value2)
  b=value2[str(year)]
  list1 = [a,b]
  plt.pie(list1,labels = [country1,country2],explode=(.2,0),colors = ("Blue","Red"))
  plt.show()
  value2 = value2.append(value1, ignore_index = True)
  value2['country'] = [country2,country1]
  value2.plot.bar(x = 'country',y=str(year),color="Orange",width=0.2)
```
